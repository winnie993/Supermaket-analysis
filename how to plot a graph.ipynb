```python
import matplotlib.pyplot as plt

# Sample data for months and income for two years
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
income_year1 = [10000, 12000, 15000, 11000, 13000, 16000, 18000, 20000, 17000, 19000, 21000, 22000]
income_year2 = [12000, 14000, 16000, 13000, 15000, 18000, 20000, 22000, 19000, 21000, 23000, 24000]

# Set the bar width
bar_width = 0.35

# Create a list of indices for each month
indices = range(len(months))

# Plotting the data using a double bar graph
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed
plt.bar(indices, income_year1, width=bar_width, label='Year 1', color='skyblue', align='center')
plt.bar([i + bar_width for i in indices], income_year2, width=bar_width, label='Year 2', color='orange', align='center')

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Income ($)')
plt.title('Monthly Income Comparison for Year 1 and Year 2')
plt.xticks([i + bar_width/2 for i in indices], months)  # Set the x-axis ticks to be the months

# Adding a legend
plt.legend()

# Adding a grid for better readability
plt.grid(axis='y')

# Display the plot
plt.show()
```
